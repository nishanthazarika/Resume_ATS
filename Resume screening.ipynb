# Clenaing data

```python
import re
```

```python
def cleaner(txt):
    txt=re.sub('https\S+\s','',txt)
    txt=re.sub('@\S+','',txt)
    txt=re.sub('#\S+','',txt)
    txt=re.sub('Rt||cc','',txt)
    txt=re.sub('[%s]' % re.escape("""!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"""),' ',txt)
    txt=re.sub('\s+',' ',txt)
    txt=re.sub(r'[^\x00-\x7f]',' ',txt)
    return txt
```

```python
cleaner('my #undb website @temple is ashttpsass/anweshadatta i love u')
```

```python
#cleaning the rsume column
```

```python
df['Resume']=df['Resume'].apply(lambda x:cleaner(x))
```

```python
df['Resume'][0]
```

# converting the categorical data into alphanumerical values


```python
df['Category']
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
le.fit(df['Category'])
df['Category']=le.transform(df['Category'])
```

```python
# converted the category to numerical corresponding
```

```python
df.Category.unique()
```

# vactorisation....

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer(stop_words='english')
tfidf.fit(df['Resume'])
reqdtext=tfidf.transform(df['Resume'])
```

```python
reqdtext[0]
```

# splitting the data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split( reqdtext,df['Category'], test_size=0.33, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

# Training the model as it is a problem of multiclassification

```python
#KneighorClassifier algorithm
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.multiclass import OneVsRestClassifier
```

```python
clf=OneVsRestClassifier(KNeighborsClassifier())
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred=clf.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```
